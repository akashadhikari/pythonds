```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.datasets import make_circles, make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN, KMeans
```

### We know K-means is effective

```python
x, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(x[:,0], x[:,1])
```

```python
kmeans = KMeans(n_clusters=4, max_iter=300)
pred_y = kmeans.fit_predict(x)
plt.scatter(x[:,0], x[:,1], c=pred_y)
plt.show()
```

### But K-means has limitations
More on StandardScaler() https://stackoverflow.com/questions/40758562/can-anyone-explain-me-standardscaler  
The idea behind StandardScaler is that it will transform your data such that its distribution will have a mean value 0 and standard deviation of 1.

```python
x, y = make_circles(n_samples=750, factor=0.3, noise=0.1)
x = StandardScaler().fit_transform(x)
plt.scatter(x[:,0], x[:,1])
```

```python
kmeans = KMeans(n_clusters=2, max_iter=300)
pred_y = kmeans.fit_predict(x)
plt.scatter(x[:,0], x[:,1], c=pred_y)
plt.show()
```

### Introducing DBSCAN

Note: `eps` is the maximum distance between two samples for one to be considered as in the neighborhood of the other.  
It specifies how close points should be to each other to be considered a part of a cluster.

```python
x, y = make_circles(n_samples=750, factor=0.3, noise=0.1)
x = StandardScaler().fit_transform(x)
y_pred = DBSCAN(eps=0.3).fit_predict(x)
plt.scatter(x[:,0], x[:,1], c=y_pred)
```
