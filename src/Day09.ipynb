## Plotting libraries

Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4,5]

y = [5,10,15,13,11,10]

plt.title("Our first graph", fontdict = {'fontname':'DejaVu Sans', 'fontsize':40, 'color': '#d08d88'}) #fontdict = {'fontname':'DejaVu Sans', 'fontsize':20}

plt.xlabel("This is x", fontdict = {'fontsize': 30})

plt.ylabel("This is y")

plt.plot(x,y,color='red') # color, linewidth

x2 = np.arange(0,6,0.5)
plt.plot(x2, x2**2)

plt.show()

```
